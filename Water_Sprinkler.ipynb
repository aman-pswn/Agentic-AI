```python
import time

# Environment
class SprinklerEnvironment:
    def __init__(self):
        self.soil = "Medium"
        self.temperature = "Hot"
        self.weather = "Sunny"
        self.sprinkler = "OFF"

    def percept(self):
        return (self.soil, self.temperature, self.weather)

    def execute(self, action):
        reward = 0

# Apply action
        if action == "TURN_ON_HIGH":
            self.sprinkler = "ON_HIGH"
            reward -= 5

            if self.soil == "Medium":
                self.soil = "High"
                reward += 20

        elif action == "TURN_ON_LOW":
            self.sprinkler = "ON_LOW"
            reward -= 2

            if self.soil == "Medium":
                self.soil = "High"
                reward += 10

        elif action == "TURN_OFF":
            self.sprinkler = "OFF"
            reward += 5

        if self.weather == "Rainy":
            self.soil = "High"
            reward += 10

        return reward

    def show(self):
        print(f"Environment -> Soil:{self.soil}, Temp:{self.temperature}, Weather:{self.weather}, Sprinkler:{self.sprinkler}")


# Agent
    def __init__(self):
        self.total_reward = 0
        self.actions = ["TURN_ON_HIGH", "TURN_ON_LOW", "TURN_OFF"]

    def utility(self, soil, temp, weather, action):
        score = 0

        if soil == "High" and action != "TURN_OFF":
            score -= 50
        if soil == "Medium":
            if action == "TURN_ON_HIGH":
                score += 40
            elif action == "TURN_ON_LOW":
                score += 25

        if temp == "Hot" and action == "TURN_ON_HIGH":
            score += 20
        if temp == "Cold" and action != "TURN_OFF":
            score -= 10

        if weather == "Rainy":
            if action == "TURN_OFF":
                score += 60
            else:
                score -= 100

        return score

# Choose best action
    def choose_action(self, percept):
        soil, temp, weather = percept

        best_action = None
        best_score = float('-inf')

        for action in self.actions:
            u = self.utility(soil, temp, weather, action)
            print(f"Utility({action}) = {u}")

            if u > best_score:
                best_score = u
                best_action = action

        return best_action

    def add_reward(self, reward):
        self.total_reward += reward


# Simulation
env = SprinklerEnvironment()
agent = SprinklerAgent()

steps = 10

print("Starting Sprinkler Simulation...\n")

for step in range(steps):
    print(f"--- Step {step+1} ---")

    percept = env.percept()
    print("Percept:", percept)

    action = agent.choose_action(percept)
    print("Chosen Action:", action)

    reward = env.execute(action)
    agent.add_reward(reward)

    env.show()
    print("Reward this step:", reward)
    print("Total Reward:", agent.total_reward)
    print()

    time.sleep(0.5)

print("Simulation Complete.")
```
